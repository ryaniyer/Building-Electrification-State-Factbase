```python
df_nat = pd.read_csv('Form 176 Type Items 1997-2019.csv',low_memory=False)
df_nat.columns = [i.replace('<BR>',' ') for i in df_nat.columns]
```

```python
def get_company_full_metrics(sa): 
    lf = pd.DataFrame(columns=['State','Company','Company Name','Metric']+[str(i) for i in range(1997,2020)])
    df = df_nat[df_nat.State==sa]
    for comp in df.Company.unique():
        print(comp)
        temp_df = df[(df.Company==comp)].filter(['Year','State','Company','Company Name']+list(df.columns[21:]))
        temp_df.index = temp_df['Year']
        for col in df.columns[21:]:
            metrics = [0 for _ in range(1997,2020)]
            for year in range(1997,2020):
                if(year in temp_df.index):
                    if(temp_df.at[year,col]>0):
                        metrics[year-1997] = temp_df.at[year, col]
            lf.loc[len(lf)] = list(temp_df.iloc[0,1:4]) + [col] + metrics
    return lf
```

```python
def get_state_metrics(sa): 
    lf = pd.DataFrame(columns=['State','Metric']+[str(i) for i in range(1997,2020)])
    df = df_nat[df_nat.State==sa]
    df.index = df['Year']
    for col in df.columns[21:]:
        metrics = [0 for _ in range(1997,2020)]
        for year in range(1997,2020):
            if(year in temp_df.index):
                if(temp_df.at[year,col]>0):
                    metrics[year-1997] = temp_df.at[year, col]
        lf.loc[len(lf)] = list(temp_df.iloc[0,1:4]) + [col] + metrics
    return lf
```

```python
def get_state_metric_totals(s):
    sa = df_nat[df_nat.State==s]
    sasum = pd.DataFrame(columns=['State','Metric']+[str(i) for i in range(1997,2020)])
    for i in sa.columns[21:]:
        metrics = [0 for _ in range(1997,2020)]
        for yr in range(1997,2020):
            temp_df = sa[sa.Year==yr]
            year_metric = np.nan_to_num(temp_df[i])
            if('Price' in i):
                try:
                    val = sum(year_metric)/np.count_nonzero(year_metric)
                    metrics[yr-1997] = round(val,2)
                except:
                    None
            else:
                metrics[yr-1997] = int(sum(year_metric))
        sasum.loc[len(sasum)] = [s,i] + list(metrics)
    return sasum
```
