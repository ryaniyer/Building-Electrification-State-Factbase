```python
import urllib.request as urllib
import zipfile
import pandas as pd
pd.set_option('display.max_columns', None)
import os
import matplotlib.pyplot as plt
```

```python
# csv parameter options:
# csv=0 --> AMI State, Counties, Cities 2018.csv
# csv=1 --> AMI Census Tracts 2018.csv
# csv=2 --> FPL State, Counties, Cities 2018.csv
# csv=3 --> FPL Census Tracts 2018.csv
# csv=4 --> SMI State, Counties, Cities 2018.csv
# csv=5 --> SMI Census Tracts 2018.csv

def get_LEAD_DataFrame(state_abbrev, csv=3):
    url = 'https://data.openei.org/files/573/' + state_abbrev + '-2018-LEAD-data.zip'
    filehandle, _ = urllib.urlretrieve(url)
    zip_file_object = zipfile.ZipFile(filehandle, 'r')
    my_file = zip_file_object.namelist()[csv]
    df = pd.read_csv(zip_file_object.open(my_file))
    zip_file_object.close()
    return df
```

```python
df = get_LEAD_DataFrame('WY', csv=3)
```

```python
df2 = df[(df['UNITS']>0.75)]
```

```python
sum(df2.UNITS)/sum(df.UNITS)
```

```python
sa = 'WY'
```

```python
df_out = pd.DataFrame({'State':[],'Tenure':[],'Year Built':[],'Structure Type':[],'Heating Fuel':[],'Federal Poverty Line':[],'Units':[],'HINCP*UNITS':[],'ELEP*UNITS':[],'GASP*UNITS':[],'FULP*UNITS':[]})
```

```python
df_out.head()
```

```python
j=0
for t in df2.TEN.unique():
    for y in df2.YBL6.unique()
        for b in df2.BLD.unique():
            for h in df2.HFL.unique():
                for f in df2.FPL15.unique():
                    if((j%100)==0):
                        print(j)
                    j+=1
                    temp_df = df2[(df2.TEN==t)&(df2.BLD==b)&(df2.YBL6==y)&(df2.HFL==h)&(df2.FPL15==f)]
                    data1 = [sa, t, y, b, h, f]
                    totunits = [sum(temp_df['UNITS']),sum(temp_df['HINCP*UNITS']),sum(temp_df['ELEP*UNITS']),sum(temp_df['GASP*UNITS']),sum(temp_df['FULP*UNITS'])]
                    df_out.loc[len(df_out.index)] = data1+totunits
```

```python
df_out['Units'].isnull().values.any()
```

```python
df3 = df_out[df_out['Units'].isnan()]
```

```python
df_out['Household Income'] = [int(df_out['HINCP*UNITS'][i]/df_out['Units'][i]) for i in df_out.index]
df_out['Electricity Payments ($)'] = [int(df_out['ELEP*UNITS'][i]/df_out['Units'][i]) for i in df_out.index]
df_out['Gas Payments ($)'] = [int(df_out['ELEP*UNITS'][i]/df_out['Units'][i]) for i in df_out.index]
df_out['Fuel Oil Payments ($)'] = [int(df_out['ELEP*UNITS'][i]/df_out['Units'][i]) for i in df_out.index]
```
